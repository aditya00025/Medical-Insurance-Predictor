```python
x = data.drop(columns='charges',axis=1)
y = data['charges']
x.head()
```

TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression  # Import Linear Regression for regression tasks
from sklearn.ensemble import RandomForestRegressor  # Import RandomForestClassifier for regression tasks
from sklearn.svm import SVR  # Import Support Vector Regression for regression tasks
from sklearn.ensemble import GradientBoostingRegressor  # Import GradientBoostingRegressor for regression tasks
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
svm = SVR()
svm.fit(X_train, y_train)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
gr = GradientBoostingRegressor()
gr.fit(X_train, y_train)
```

```python
xgb_model = xgb.XGBRegressor(
    objective ='reg:squarederror',  # Objective for regression
    n_estimators=100,                # Number of trees (estimators)
    learning_rate=0.1,               # Learning rate
    max_depth=5,                     # Maximum depth of trees
    random_state=42                  # For reproducibility
)
xgb_model.fit(X_train, y_train)
```

```python
y_pred1 = lr.predict(X_test)

# Make predictions using the Support Vector Regression (SVR) model on the test data
y_pred2 = svm.predict(X_test)

# Make predictions using the Random Forest Regressor (RF) model on the test data
y_pred3 = rf.predict(X_test)

# Make predictions using the Gradient Boosting Regressor (GR) model on the test data
y_pred4 = gr.predict(X_test)

# Make predictions using the XGBoost model on the test data
y_pred5 = xgb_model.predict(X_test)

# DataFrame to compare actual and predicted values
df1 = pd.DataFrame({'Actual': y_test, 'LR': y_pred1, 'SVR': y_pred2, 'RF': y_pred3, 'GR': y_pred4,'XGB':y_pred5})
df1
```

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(15, 10))

# Linear Regression Predictions
plt.subplot(231)
plt.plot(df1['Actual'].iloc[0:11], label='Actual', color='blue')
plt.plot(df1['LR'].iloc[0:11], label="LR", color='orange')
plt.title('Linear Regression Predictions', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid()

# Support Vector Regression Predictions
plt.subplot(232)
plt.plot(df1['Actual'].iloc[0:11], label='Actual', color='blue')
plt.plot(df1['SVR'].iloc[0:11], label="SVR", color='green')
plt.title('Support Vector Regression Predictions', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid()

# Random Forest Predictions
plt.subplot(233)
plt.plot(df1['Actual'].iloc[0:11], label='Actual', color='blue')
plt.plot(df1['RF'].iloc[0:11], label="RF", color='red')
plt.title('Random Forest Predictions', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid()

# Gradient Boosting Predictions
plt.subplot(234)
plt.plot(df1['Actual'].iloc[0:11], label='Actual', color='blue')
plt.plot(df1['GR'].iloc[0:11], label="GR", color='purple')
plt.title('Gradient Boosting Predictions', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid()

# XGBoost Predictions
plt.subplot(235)
plt.plot(df1['Actual'].iloc[0:11], label='Actual', color='blue')
plt.plot(df1['XGB'].iloc[0:11], label="XGB", color='brown')
plt.title('XGBoost Predictions', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.grid()

# Adjust layout
plt.tight_layout()

# Show the entire plot
plt.show()
```

```python
r2_score_array = []

# Calculate and store the R-squared (R2) score for the Linear Regression (LR) model
lr_r2_score = r2_score(y_test, y_pred1)
r2_score_array.append({'score':lr_r2_score,'model':'LinearRegression'})

# Calculate and store the R-squared (R2) score for the Support Vector Regression (SVR) model
svm_r2_score = r2_score(y_test, y_pred2)
r2_score_array.append({'score':svm_r2_score,'model':'SVM'})


# Calculate and store the R-squared (R2) score for the Random Forest Regressor (RF) model
rf_r2_score = r2_score(y_test, y_pred3)
r2_score_array.append({'score':rf_r2_score,'model':'RandomForest'})


# Calculate and store the R-squared (R2) score for the Gradient Boosting Regressor (GR) model
gr_r2_score = r2_score(y_test, y_pred4)
r2_score_array.append({'score':gr_r2_score,'model':'GradientBoosting'})


# Calculate and store the R-squared (R2) score for the XGBoost model
xgb_r2_score = r2_score(y_test, y_pred5)
r2_score_array.append({'score':xgb_r2_score,'model':'Extreme gradient boost'})

r2_score_array = np.array(r2_score_array)
```

```python
print("R-squared Score for Linear Regression (LR):", lr_r2_score)
print("R-squared Score for Support Vector Regression (SVR):", svm_r2_score)
print("R-squared Score for Random Forest Regressor (RF):", rf_r2_score)
print("R-squared Score for Gradient Boosting Regressor (GR):", gr_r2_score)
print("R-squared Score for XGBoost:", xgb_r2_score)
max_score_model = max(r2_score_array, key=lambda x: x['score'])

print('The winner is : ',max_score_model['model'],'\nwith a score of : ',max_score_model['score'])
```

```python
# Calculate and store the Mean Absolute Error (MAE) for the Linear Regression (LR) model
mae_lr = mean_absolute_error(y_test, y_pred1)

# Calculate and store the Mean Absolute Error (MAE) for the Support Vector Regression (SVR) model
mae_svr = mean_absolute_error(y_test, y_pred2)

# Calculate and store the Mean Absolute Error (MAE) for the Random Forest Regressor (RF) model
mae_rf = mean_absolute_error(y_test, y_pred3)

# Calculate and store the Mean Absolute Error (MAE) for the Gradient Boosting Regressor (GR) model
mae_gr = mean_absolute_error(y_test, y_pred4)

# Calculate and store the Mean Absolute Error (MAE) for the XGBoost model
mae_xgb = mean_absolute_error(y_test, y_pred5)

mae_array = [{
    'model':'LinearRegression',
    'mae':mae_lr
},{
    'model':'Support Vector Regression',
    'mae':mae_svr
},{
    'model':'Random Forest Regressor',
    'mae':mae_rf
},{
    'model':'Gradient Boosting Regressor',
    'mae':mae_gr
},{
    'model':'XGBoost',
    'mae':mae_xgb
}
]
```

```python
print("MAE for Linear Regression (LR):", mae_lr)
print("MAE for Support Vector Regression (SVR):", mae_svr)
print("MAE for Random Forest Regressor (RF):", mae_rf)
print("MAE for Gradient Boosting Regressor (GR):", mae_gr)
print("MAE for XGBoost:", mae_xgb)

min_score_model = min(mae_array, key=lambda x: x['mae'])

print('The winner is : ',min_score_model['model'],'\nwith a MAE of : ',min_score_model['mae'])
```
